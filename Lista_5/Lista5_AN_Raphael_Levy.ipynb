```python
import numpy as np
import matplotlib.pyplot as plt
```

### b)

```python
def BTCS(c, d, L, T, delta_t, delta_x): 
    
    # Supõe que T/dt e L/dx são inteiros 
    N_t = int(T/delta_t)
    N_x = int(L/delta_x)
    v = c*delta_t/(delta_x)**2
    
    ones = np.ones(N_x - 2)
    M = (2*v + 1 - d*delta_t)*np.eye(N_x-1) -v*(np.diag(ones, -1) + np.diag(ones, 1)) 
        # 2v +1 - d\Delta t na diagonal principal; -v na diagonal superior e inferior
    
    U = np.zeros((N_x+1, N_t+1))
    U[:,0] = [(np.sin(np.pi/2 * i * delta_x))**2 for i in range(N_x+1)]
    for j in range(1, N_t):
        U[1:N_x, j] = np.linalg.solve(M, U[1:N_x,j-1])
        
        
    # Plotando:     
    x = np.arange(0, L+delta_x, delta_x)
    t = np.arange(0, T+delta_t, delta_t)
    X, Y = np.meshgrid(t, x)
    
    # Plot 3D
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    ax.contour3D(X, Y, U, 50)
    ax.set_xlabel('t')
    ax.set_ylabel('x')
    ax.set_zlabel('u')
    plt.show()
```

Testando o código, comparando valores de $\Delta_t$ e $\Delta_x$

```python
BTCS(c=1, d=1, L=1, T=1, delta_t=0.1, delta_x=0.1)
```

```python
BTCS(c=1, d=1, L=1, T=1, delta_t=0.01, delta_x=0.01)
```
